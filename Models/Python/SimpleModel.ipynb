```python
import numpy as np
import numpy_financial as npf
```

### Definition of inputs

```python
salary = 60000
savings_rate = 0.25
investment_rate = 0.05
desired_cash = 1500000
```

### Determining cash saved

```python
annual_cash = salary*savings_rate
annual_cash
```

### Years to retirement

```python
years_to_retirement = npf.nper(investment_rate, -annual_cash, 0, desired_cash)
```

```python
print(f'Pedro has {years_to_retirement:.1f} years to retirement')
```

### Multiple Interest Rates

```python
investment_rates = [0.05, 0.06, 0.07]
```

```python
for i_rate in investment_rates:
    years_to_retirement = npf.nper(i_rate, -annual_cash, 0, desired_cash)
    print(f'Pedro has {years_to_retirement:.1f} years to retirement at a {i_rate:.0%} return')
```

### Multiple Rates (Interest and Savings)

```python
investment_rates = [0.05, 0.06, 0.07]
savings_rates = [0.1, 0.25, 0.4]
```

```python
for i_rate in investment_rates:
    for s_rate in savings_rates:
        years_to_retirement = npf.nper(i_rate, -annual_cash, 0, desired_cash)
        print(f'Pedro has {years_to_retirement:.1f} years to retirement at a {i_rate:.0%} return')
```
